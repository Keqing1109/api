# Exercise 1: Creating Property Listings with Python Dictionaries


## Step 1: Define Sample Property Listings

Each property should include:
- `property_id`
- `location`
- `type`
- `price_per_night`
- `features` (list of strings)
- `tags` (list of strings)


```python
# Define 5 property dictionaries
property1 = {
    "property_id": 1,
    "location": "Lake Muskoka",
    "type": "cabin",
    "price_per_night": 180,
    "features": ["fireplace", "canoe", "bbq"],
    "tags": ["lakefront", "family-friendly", "nature"]
}

property2 = {
    "property_id": 2,
    "location": "Toronto Downtown",
    "type": "condo",
    "price_per_night": 220,
    "features": ["wifi", "parking", "gym"],
    "tags": ["urban", "nightlife", "business"]
}

property3 = {
    "property_id": 3,
    "location": "Blue Mountain",
    "type": "chalet",
    "price_per_night": 250,
    "features": ["ski-in", "hot tub", "fireplace"],
    "tags": ["skiing", "luxury", "mountains"]
}

property4 = {
    "property_id": 4,
    "location": "Niagara-on-the-Lake",
    "type": "house",
    "price_per_night": 160,
    "features": ["garden", "bbq", "patio"],
    "tags": ["romantic", "wine", "quiet"]
}

property5 = {
    "property_id": 5,
    "location": "Prince Edward County",
    "type": "cottage",
    "price_per_night": 190,
    "features": ["kayak", "beach access", "bbq"],
    "tags": ["beachfront", "family-friendly", "pets"]
}
```

```python
import random
#It can also be different types of probability, not random, deep learning time serise detect fraud, nuilt the stimulatir to generate the case situations, create the anomolies.
#Real situations should use the probaility settings
LOCATIONS = [
    "Lake Muskoka", "Toronto Downtown", "Blue Mountain",
    "Niagara-on-the-Lake", "Prince Edward County", "Collingwood",
    "Wasaga Beach", "Kingston", "Ottawa", "Halifax"
]

TYPES = ["cabin", "condo", "chalet", "house", "cottage", "loft", "townhouse"]

FEATURES_POOL = [
    "wifi", "parking", "gym", "hot tub", "fireplace", "bbq",
    "patio", "garden", "ski-in", "beach access", "canoe", "kayak",
    "air conditioning", "washer", "dryer"
]

TAGS_POOL = [
    "lakefront", "beachfront", "family-friendly", "pets", "luxury",
    "urban", "nightlife", "business", "mountains", "romantic", "quiet", "nature"
]

def make_random_property(property_id):
    location = random.choice(LOCATIONS)
    ptype = random.choice(TYPES)
    price = random.randint(120, 350)  # tweak range as needed
#sample function, use the feature pool and the numbers two to four.Demos counsultant can be replced by LLM.
    features = random.sample(FEATURES_POOL, k=random.randint(2, 4))
    tags = random.sample(TAGS_POOL, k=random.randint(2, 3))
    return {
        "property_id": property_id,
        "location": location,
        "type": ptype,
        "price_per_night": price,
        "features": features,
        "tags": tags,
    }

```

## Step 2: Store Properties in a List


```python
# Option 1: Store properties in a list
property_listings = [property1, property2, property3, property4, property5]
```

```python
# Option 2: generate 5 random properties
random.seed(42)  # optional: for reproducible results
#seeds are setted in compute and default unchanged, make_random_property is a dictionary abstract randomly.
property_listings = [make_random_property(i) for i in range(1, 6)]

for prop in property_listings:
    print(prop)
```

## Step 3: Print All Listings in Readable Format

Use a loop to print each listing.


```python
# Print all listings
for prop in property_listings:
    print(f"Property ID: {prop['property_id']}")
    print(f"Location: {prop['location']}")
    print(f"Type: {prop['type']}")
    print(f"Price per night: ${prop['price_per_night']}")
    print(f"Features: {', '.join(prop['features'])}")
    print(f"Tags: {', '.join(prop['tags'])}")
    print("-" * 40)
```

## Step 4: Filter Listings by Budget

Print only properties under a given budget.


```python
# Filter properties under a given budget
budget = 200
print(f"Properties under ${budget} per night:")
print("=" * 40)
for prop in property_listings:
    if prop["price_per_night"] <= budget:
        print(f"{prop['location']} - ${prop['price_per_night']}")
```

## Bonus: Search by Tag

Let the user input a tag (e.g., `beachfront`, `family-friendly`) and print matching properties.


```python
# Search by tag
#Faozy matching from somebody write
search_tag = input("Enter a tag to search (e.g., 'beachfront'): ").strip().lower()
print(f"Properties tagged with '{search_tag}':")
print("=" * 40)
for prop in property_listings:
#search tags in the pool of tags
    if search_tag in [tag.lower() for tag in prop["tags"]]:
        print(f"{prop['property_id']} - {prop['location']} - Tags: {', '.join(prop['tags'])}")
#json homewrok will be done with TA on Monday.
```
