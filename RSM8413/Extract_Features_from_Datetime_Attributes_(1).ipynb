```python
import pandas as pd
import numpy as np
df = pd.read_csv('Traffic data.csv', nrows=100)
df.head()
```

```python
# Remove additional Atrributes that are not needed for us before feature engineering

df = df.drop(columns = ['ID', 'Count'])
```

```python
df['Datetime'] = pd.to_datetime( df['Datetime']) # Using the Default function in Pandas "To date Time"
df.head()

# You can see that before we didn't has the secs data, now, is having minutes and second, formate also change into the year month date
```

## Extrate date features

```python
df['year'] = df['Datetime'].dt.year # Create a new column variable in dataframe, dt represent datetime funtion, now we go into and extract what we want inplace
df['month'] = df['Datetime'].dt.month
df['day'] = df['Datetime'].dt.day
df['quarter'] = df['Datetime'].dt.quarter
df['day_of_week'] = df['Datetime'].dt.dayofweek
#df['week'] = df['Datetime'].dt.week
df['is_weekend'] = np.where(df['day_of_week'].isin([5,6]),1,0)# day of week starting from 0,6
df.head()
df.sample(frac=1).head(100) # shuffling
```

## Extract Time features/attributes

```python
df['hour'] =df['Datetime'].dt.hour
df['minute'] =df['Datetime'].dt.minute
df['second'] =df['Datetime'].dt.second
df.sample(frac=1).head()
```

```python
df['date'] =df['Datetime'].dt.date
df['time'] =df['Datetime'].dt.time
df.head()
```

```python

import datetime
## find difference from currect day
df['difference'] = datetime.datetime.today()-df['Datetime']
df.head()
```
